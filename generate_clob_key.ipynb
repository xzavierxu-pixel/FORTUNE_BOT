# Generate CLOB API Keys

This notebook helps you create or derive your CLOB API credentials from your private key.

It uses the `py_clob_client` located in the `py-clob-client/` directory.

```python
import os
import sys
from pathlib import Path

# Ensure we can import py_clob_client
repo_root = Path(os.getcwd())
clob_client_path = repo_root / "py-clob-client"
if str(clob_client_path) not in sys.path:
    sys.path.append(str(clob_client_path))

try:
    from py_clob_client.client import ClobClient
    from py_clob_client.clob_types import ApiCreds
    print("Successfully imported ClobClient")
except ImportError as e:
    print(f"Error importing py_clob_client: {e}")
    print("Make sure py-clob-client/ exists and dependencies are installed.")
    raise
```

```python
# Configuration
HOST = "https://clob.polymarket.com"
CHAIN_ID = 137 # Polygon

# It is recommended to set PEG_CLOB_PRIVATE_KEY in your environment or a .env file
#PRIVATE_KEY = os.getenv("PEG_CLOB_PRIVATE_KEY") 
PRIVATE_KEY = ""
if not PRIVATE_KEY:
    # Replace with your actual private key for manual run, 
    # but BE CAREFUL not to commit it!
    PRIVATE_KEY = "YOUR_PRIVATE_KEY_HERE"
    
if PRIVATE_KEY == "YOUR_PRIVATE_KEY_HERE":
    print("Please provide a valid private key in the code or set PEG_CLOB_PRIVATE_KEY environment variable.")
else:
    print("Private key found. Initializing client...")
    client = ClobClient(
        host=HOST,
        key=PRIVATE_KEY,
        chain_id=CHAIN_ID
    )
    
    try:
        print("Attempting to create/derive API key...")
        creds = client.create_or_derive_api_creds()
        
        if creds:
            print("\nSuccess! CLOB API Credentials generated:\n")
            print(f"PEG_CLOB_API_KEY={creds.api_key}")
            print(f"PEG_CLOB_API_SECRET={creds.api_secret}")
            print(f"PEG_CLOB_API_PASSPHRASE={creds.api_passphrase}")
        else:
            print("Failed to generate credentials (returned None).")
            
    except Exception as e:
        print(f"An error occurred: {e}")
```
